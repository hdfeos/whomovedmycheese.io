This is Amazon Sidewalk data processing example.

```python
import pandas as pd

from google.colab import files

uploaded = files.upload()


```

```python
for fn in uploaded.keys():
  df = pd.read_csv(fn)
```

```python
display(df)
```

```python
from numpy import NaN
 # Filter NaN and filter by date
df_n = df[df.GPS_latitude.notnull() & (df.timestamp < '2023-04-07 00:00:00')]
display(df_n)
```

```python
# Sort data by time.
df_s = df_n.sort_values('timestamp')
display(df_s)
```

```python
from pandas.core.internals.base import default_index
import datetime
s = pd.to_datetime(df_n['timestamp']).astype(int) / 10**9
display(s)

import h5py
with h5py.File('RAT_demo.h5', 'w') as f:
            f['Time'] = s
            f['Time'].attrs['units'] = "seconds since 1970-01-01 00:00:00"
            f['Latitude'] = df_n.GPS_latitude
            f['Latitude'].attrs['units'] = "degrees_north"
            f['Longitude'] = df_n.GPS_longitude
            f['Longitude'].attrs['units'] = "degrees_east"
            f['Altitude'] = df_n.GPS_accuracy
            f['Altitude'].attrs['units'] = "m"
            f['LQI'] = df_n.LQI #  Link Quality Indication
            
            # Creating dimension scales
            h5py.h5ds.set_scale(f['Time'].id)

            f['Altitude'].dims[0].attach_scale(f['Time'])

            f['Latitude'].dims[0].attach_scale(f['Time'])

            f['Longitude'].dims[0].attach_scale(f['Time'])

            f['LQI'].dims[0].attach_scale(f['Time'])
```

```python
files.download('RAT_demo.h5')
```
